## Covid Data

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from datetime import date, datetime, time, timezone

df = pd.read_csv("./Data/owid-covid-data.csv")
```

```python
df.info()
```

```python
df.set_index("date", inplace=True)
df.head()
```

```python
len(df["location"].unique())
```

```python
covid_c = df.groupby(["location"])
```

```python
# Separate the information of the created group by key and print them.
for key, group in covid_c:
    print("+key", key)
    print("+number", len(group))
    print(group.head())
    print("\n")
```

```python
total_df = covid_c.sum()
total_df.head()
```

```python
import folium
import json
```

```python
center =  [40.43961762440417, -3.697316046199983]
```

```python
map = folium.Map(location=center,
               zoom_start=15,
               tiles= "stamenwatercolor")
map
```

```python
marker_map = folium.Map(center, zoom_start=15, tiles="Stamen Terrain")

folium.Marker(location=[40.444300753764196, -3.706115648828032],
              popup="Ocio y Deporte Canal",
              icon=folium.Icon(color="orange"),
              ).add_to(marker_map)

marker_map
```

```python
url = ("./Data/World_Countries_Generalized.geojson")
json_data = json.load(open(url), encoding="utf-8")

state_data = total_df["total_cases"]
state_data
```

```python
m = folium.Map(location=center, zoom_start=4)

folium.Choropleth(geo_data=json_data, 
                  name="choropleth",
                  data=total_df,
                  columns=["location", "total_cases"],
                  ).add_to(m)

folium.LayerControl().add_to(m)
```

```python
m = folium.Map(location=center, zoom_start=2,
               max_bounds=True,
               min_zoom=1, min_lat=-84,
               max_lat=84, min_lon=-175, max_lon=187)
m
```
